```python
start = 1276311666.029
df = pd.read_csv("flowfile.csv",low_memory=False)
```

```python
df.head(5)
```

```python
df["Label"] = "Background Flow"
```

```python
attacks = []
for line in open("datasets/attacks.txt").readlines():
    line = line.rstrip().split()
    if attacks:
        attacks[-1].append(int(line[5]))
    attacks.append([line[0], line[2], int(line[5])])

attacks[-1].append(1440)
    
for attack in attacks:
    df.loc[(df["DstIP"] == attack[1]) & 
       (df["StartTime"] > start + (attack[2]*60)) & 
       (df["EndTime"] < start + (attack[3]*60)) &
       ((df["SrcIP"] == "192.168.56.1") | (df["SrcIP"] == "192.168.56.102")),"Label"] = attack[0]    
    
# pprint(attacks)
```
